Usando los métodos estudiados, determine todas las raíces de la ecuación:

$x^5+5.5x^4+4.5x^3-9.5x^2-15.5x-6 = 0$

```python
from scipy import optimize

def f(x):
    return x**5 + 5.5*x**4 + 4.5*x**3 - 9.5*x**2 - 15.5*x - 6

rangos = [(-6, -4),(-3, -1), (0,2)]

for r in rangos:
    raiz = optimize.bisect(f, r[0], r[1])
    print(f"Raíz en el intervalo {r}: {raiz}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)
y = f(x)

plt.plot(x, y, label='f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Gráfico de la función f(x)')
plt.xlabel('x') 
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.show()
```
